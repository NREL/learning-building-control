```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
rtp = pd.read_csv('real_time_price.csv')
```

```python
rtp['Datetime'] = pd.to_datetime(rtp['Datetime'])
rtp.head()
```

```python
start_date = pd.to_datetime('2020-07-01 00:00:00')
end_date = pd.to_datetime('2020-07-08 00:00:00')

rtp_select = rtp[(rtp['Datetime'] > start_date) & (rtp['Datetime'] < end_date)]
rtp_select.set_index('Datetime', inplace=True)
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
rtp_select.DAP.plot()
rtp_select.RTP.plot()
plt.legend()
```
